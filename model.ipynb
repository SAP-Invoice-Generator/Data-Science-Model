```python
import pandas as pd

# Load the original DataFrame from the GitHub repository
url = 'https://raw.githubusercontent.com/SAP-Invoice-Generator/Data-Science-Model/main/DATA.csv'
df_original = pd.read_csv(url)

# Define the duplicate dataset
duplicate_data = {
    'invoice_number': ['INV004', 'INV005'],
    'date': ['2024-03-24', '2024-03-25'],
    'total amount': [120.0, 180.0],
    'currency type': ['USD', 'EUR'],
    'no of items': [6, 8],
    'customer name': ['Bob Smith', 'Emily Johnson']
}

# Create DataFrame from the duplicate data
df_duplicate = pd.DataFrame(duplicate_data)

# Concatenate the duplicate dataset to the original DataFrame
df_updated = pd.concat([df_original, df_duplicate], ignore_index=True)

```

```python
df_updated
```

Updating to CSV File

```python
df_updated.to_csv('DATA.csv', index=False)
```

Data Preprocessing

DATE FORMATTING

```python
df_updated['date'] = pd.to_datetime(df_updated['date'], errors='coerce').dt.strftime('%Y-%m-%d')
```

```python
df_updated
```

CURRENCY TYPE FORMATTING

```python
df_updated['currency type'] = df_updated['currency type'].replace({'$': 'USD', '€': 'EUR', '£': 'GBP'})
```

```python
df_updated['currency type'] = df_updated['currency type'].str.replace('[$]', 'USD')
df_updated['currency type'] = df_updated['currency type'].str.replace('[€]', 'EUR')
df_updated['currency type'] = df_updated['currency type'].str.replace('[£]', 'GBP')
```

```python
df_updated
```

Making non currency values as NULL

```python
import numpy as np

# Define a function to check if a value is a number
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False

# Replace any numbers with NULL values
df_updated['currency type'] = df_updated['currency type'].apply(lambda x: np.nan if is_number(x) else x)
```

```python
df_updated
```

Total Amount is preprocessed

```python
df_updated['total amount'] = pd.to_numeric(df_updated['total amount'], errors='coerce')
```

```python
df_updated
```

Total amount is an important feature for us so data with no amount is removed from the table

```python
df_updated.dropna(subset=['total amount'], inplace=True)
```

```python
df_updated
```

Preprocessing Customer Name column

```python
# Replace exact matches of 'none', 'null', or 'None' with 'NULL'
df_updated['customer name'] = df_updated['customer name'].str.strip().replace(['none', 'null', 'None'], 'NULL')
```

```python
df_updated
```

PLOTTING BAR GRAPH BETWEEN MONTH VS TOTAL AMOUNT

```python
import matplotlib.pyplot as plt
import numpy as np

# Define conversion rates
usd_to_inr_rate = 83
eur_to_inr_rate = 90

# Convert 'total amount' column from USD to INR using the conversion rate
df_updated.loc[df_updated['currency type'] == 'USD', 'total amount'] *= usd_to_inr_rate
df_updated.loc[df_updated['currency type'] == 'USD', 'currency type'] = 'INR'

# Convert 'total amount' column from EUR to INR using the conversion rate
df_updated.loc[df_updated['currency type'].isin(['EUR', 'Euro']), 'total amount'] *= eur_to_inr_rate
df_updated.loc[df_updated['currency type'].isin(['EUR', 'Euro']), 'currency type'] = 'INR'
```

```python
# Group the DataFrame by month and calculate the sum of 'total amount'
df_monthly_total = df_updated.groupby(df_updated['date'].dt.month)['total amount'].sum()

# Extracting month and total amount data
months_with_data = df_monthly_total.index
total_amounts = df_monthly_total.values
```

```python
# Creating an array of all months from 1 to 12
all_months = np.arange(1, 13)

# Creating a dictionary to store total amounts for months
monthly_dict = dict(zip(months_with_data, total_amounts))

# Creating an array to store total amounts for all months
all_total_amounts = [monthly_dict.get(month, 0) for month in all_months]
```

```python
# Plotting the bar chart
plt.bar(all_months, all_total_amounts)

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Total Amount (INR)')
plt.title('Month-wise Total Amount')

# Displaying the plot
plt.show()
```

```python
# Extracting year from the 'date' column
df_updated['year'] = df_updated['date'].dt.year

# Group the DataFrame by year and calculate the sum of 'total amount'
df_yearly_total = df_updated.groupby('year')['total amount'].sum()

# Extracting years and total amount data
years_with_data = df_yearly_total.index
total_amounts = df_yearly_total.values

# Creating an array of all years from the minimum to maximum year in the data
all_years = np.arange(min(years_with_data), max(years_with_data) + 1)

# Creating a dictionary to store total amounts for years
yearly_dict = dict(zip(years_with_data, total_amounts))

# Creating an array to store total amounts for all years
all_total_amounts = [yearly_dict.get(year, 0) for year in all_years]
```

```python
# Plotting the bar chart
plt.bar(all_years, all_total_amounts)

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Total Amount (INR)')
plt.title('Year-wise Total Amount')

# Displaying the plot
plt.show()
```

```python
# Plotting the line chart
plt.plot(years_with_data, total_amounts, marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Total Amount (INR)')
plt.title('Year-wise Total Amount')

# Displaying the plot
plt.grid(True)
plt.show()
```
