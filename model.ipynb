```python
import subprocess
import pandas as pd

# Load the original DataFrame from the GitHub repository
url = 'https://raw.githubusercontent.com/SAP-Invoice-Generator/Data-Science-Model/main/DATA.csv'
df_original = pd.read_csv(url)

# Define the duplicate dataset
duplicate_data = {
    'invoice_number': ['INV004', 'INV005'],
    'date': ['2024-03-24', '2024-03-25'],
    'total amount': [120.0, 180.0],
    'currency type': ['USD', 'EUR'],
    'no of items': [6, 8],
    'customer name': ['Bob Smith', 'Emily Johnson']
}

# Create DataFrame from the duplicate data
df_duplicate = pd.DataFrame(duplicate_data)

# Append the duplicate dataset to the original DataFrame
df_updated = df_original.append(df_duplicate, ignore_index=True)

```
